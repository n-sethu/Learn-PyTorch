```python
import torch 
import torchvision
import numpy as np
```

```python
# tensors 
data =[[1,2],[3,4]]
x_data = torch.tensor(data)
scalar = 7
t_scale = torch.tensor(scalar)
print(x_data,t_scale)
print(x_data.ndim)

# using np array
np_arr = np.array(data)
x_np = torch.from_numpy(np_arr)
print(x_np)

# from another tensor args (shape, dtype)
x_same = torch.ones_like(x_data)

x_rand = torch.rand_like(x_data,dtype=torch.float)
print(x_rand)
```

```python
# shape is a tuple of tensor dimensions
shape = (2,3)
rand_tensor = torch.rand(shape)
ones_tensor = torch.ones(shape)
zeros_tensor = torch.zeros(shape)

print(f"Random Tensor: \n {rand_tensor} \n")
print(f"Ones Tensor: \n {ones_tensor} \n")
print(f"Zeros Tensor: \n {zeros_tensor}")
```

```python
# Has attributes shape, dtype and device it is stored on
tensor = torch.rand((2,3,2))
print(f"Shape of tensor: {tensor.shape}")
print(f"Datatype of tensor: {tensor.dtype}")
print(f"Device tensor is stored on: {tensor.device}")
```

```python
# Operations
torch.backends.mps.is_available()
if torch.accelerator.is_available():
    tensor=tensor.to(torch.accelerator.current_accelerator())
    
```

```python
# NumPy like indexing/slicing
tensor = torch.tensor([[1,3,5],[1,5,6],[0,9,4],[31,3,5]])
print(f"First Row: {tensor[0]}")
print(f"First Column: {tensor[:,0]}")
print(f"Last Column: {tensor[:,-1]}")
```

```python
tensor[:,1]=0 #make the 2nd column zeros 
tensor
```

```python
# torch.cat to join tensors (extends along one dimension)
t1 = torch.cat([tensor,tensor,tensor],dim=0)
print(t1)
# can also use torch.stack (creates a new dimension (2d->3d or 3d->4d))
torch.stack([tensor,tensor,tensor],dim=0)
```

```python
y1 = t1 @ t1.T
print(y1)
y2 = tensor.matmul(tensor.T)
print(y2)

z1= tensor * tensor
z1.add_(5)
```

```python
# tensor to np
t = torch.ones(5)
print(f"t: {t}")
n= t.numpy()
n
```

```python
# they share the same memory sooo
t.add_(3.14)
n
```

```python
n = np.ones(5)
t= torch.from_numpy(n)
t
```

```python
# works both ways
np.add(n,3.17,out=n)
print(f"t: {t}")
print(f"n: {n}")
```
