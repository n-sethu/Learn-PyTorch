```python
tensor_ops = torch.ones(2,3,4)
print(tensor_ops)
tensor_ops = tensor_ops *10
print(tensor_ops)
tensor_ops = tensor_ops - torch.multiply(tensor_ops,2)
print(tensor_ops)
```

```python
# Matrix Multiplication (finally)
# inner dimensions must match
t3 = torch.ones(2,4,4)
t3=t3+3
print(torch.matmul(tensor_ops,t3))
t4 = torch.kron(tensor_ops,t3)
print(t4.shape)
```
